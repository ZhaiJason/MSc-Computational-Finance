## Fourier Transform of a Gaussian

```python
import numpy as np
from numpy.fft import fft, ifft, fftshift, ifftshift
import matplotlib.pyplot as plt

import numpy as np
import matplotlib.pyplot as plt

# Define parameters
N = 1024
Dx = 0.1
x = Dx * np.arange(-N/2, N/2)
L = N * Dx
Dxi = 2 * np.pi / L # Nyquist relation: Dx*Dxi = 2*pi/N
xi = Dxi * np.arange(-N/2, N/2)

# Gaussian in x space and its Fourier transform
a = 1
fa = np.exp(-a * x**2) # Gaussian in x space
Fa = np.sqrt(np.pi/a) *np.exp(-xi**2 / (4*a)) # Gaussian in xi space

# Fourier transforms
Fn = np.fft.fftshift(np.fft.ifft(np.fft.ifftshift(fa))) * L
fn = np.fft.fftshift(np.fft.fft(np.fft.ifftshift(Fa))) / L
Fn1 = np.fft.fftshift(np.fft.fft(np.fft.ifftshift(fa))) * Dx
fn1 = np.fft.fftshift(np.fft.ifft(np.fft.ifftshift(Fa))) / Dx

plt.figure(1)
plt.plot(x, np.real(fn),  linestyle='-.', color="r", label='Re(fn)')
plt.plot(x, np.imag(fn),  linestyle=':', color="g", label='Im(fn)')
plt.plot(x, fa, linestyle='solid', color="y", label='fa')
plt.xlabel('xi')
plt.ylabel('F')
plt.legend()
plt.axis([-5, 5, 0, 1])
plt.show()

plt.figure(2)
plt.plot(xi, np.real(Fn),  linestyle='-.', color="r", label='Re(Fn)')
plt.plot(xi, np.imag(Fn),  linestyle=':', color="g", label='Im(Fn)')
plt.plot(xi, Fa, linestyle='solid', color="y", label='Fa')
plt.xlabel('xi')
plt.ylabel('F')
plt.legend()
plt.axis([-5, 5, 0, 1])
plt.show()

plt.figure(3)
plt.plot(x, np.real(fn1),  linestyle='-.', color="r", label='Re(fn1)')
plt.plot(x, np.imag(fn1),  linestyle=':', color="g", label='Im(fn1)')
plt.plot(x, fa, linestyle='solid', color="y", label='fa')
plt.xlabel('xi')
plt.ylabel('F')
plt.legend()
plt.axis([-5, 5, 0, 1])
plt.show()

plt.figure(4)
plt.plot(xi, np.real(Fn1),  linestyle='-.', color="r", label='Re(Fn1)')
plt.plot(xi, np.imag(Fn1),  linestyle=':', color="g", label='Im(Fn1)')
plt.plot(xi, Fa, linestyle='solid', color="y", label='Fa')
plt.xlabel('xi')
plt.ylabel('F')
plt.legend()
plt.axis([-5, 5, 0, 1])
plt.show()
```
